```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
```

```python
rf=RandomForestRegressor(n_estimators=100)
svr=SVR(kernel='rbf')
lr=LinearRegression()
```

```python
X=df.iloc[:,:-1].values
Y=df.iloc[:,-1].values
```

```python
len(X),len(Y)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
X_train
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

```python
def model_train(X_train, X_test, Y_train, Y_test,model,model_name):
    model.fit(X_train,Y_train)
    y_pred=model.predict(X_test)
    mse=mean_squared_error(Y_test,y_pred=y_pred)
#     return mse
#     acc=accuracy_score(Y_test,y_pred)
    
    print(model_name,mse)
    
```

```python
%%timeit
model_train(X_train,X_test,Y_train,Y_test,lr,"Linear Regression")
```

```python
%%timeit
model_train(X_train,X_test,Y_train,Y_test,svr,"SVR")
```

```python
%%timeit
model_train(X_train,X_test,Y_train,Y_test,rf,"Random Forest")
```

```python
rf.fit(X_train,Y_train)
y_pred=rf.predict(X_test)
mse=mean_squared_error(Y_test,y_pred=y_pred)
```
